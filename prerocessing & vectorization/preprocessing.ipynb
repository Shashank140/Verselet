```python
poems_data_path = '/poems_data_merged.csv'
```

```python
stop_words_path = 'stopwords2.txt'
poems_data = open(poems_data_path,'r',errors='ignore')
stop_words_file = open("stopwords2.txt",'r')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
reader = csv.reader(poems_data, delimiter=',')
```

```python
stop_words = []
for stop_word in stop_words_file:
	stop_words.append(stop_word)
```

```python
line_count = 0

all_data = []
```

```python
for row in reader:
	if (line_count == 0):
		pass
	else:
		tokenized_poem = tokenize(row[4],'hi')
		filtered_poem = [word for word in tokenized_poem if not word in stop_words] 
		all_data.append([row[0],row[1],row[2],row[3],filtered_poem])
	line_count += 1
```

```python
for i in range (80) :
	print(all_data[i])
```

```python
with open('processed_data3.txt', 'wb') as processed_data3:
   pickle.dump(all_data, processed_data3)

  #  stopwords1.txt ->processed_data2.txt
  #  stopwords2.txt ->processed_data3.txt
  # stopwords3.txt ->processed_data.txt
```
